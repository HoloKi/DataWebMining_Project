```python
import statsmodels.api as sm
lm = sm.OLS.from_formula('Sale_Price ~ Gr_Liv_Area', df)
result= lm.fit()
result.summary()
```

```python
lm1 = sm.OLS.from_formula('Sale_Price ~ Gr_Liv_Area+ Year_Built', df)
lm1.fit().summary()
```
